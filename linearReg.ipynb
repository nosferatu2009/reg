```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

def calc_coeff(x,y) :
    n = np.size(x)
    mx = np.mean(x)
    my = np.mean(y)
    
    cd = np.sum(x*y)-n*mx*my
    d = np.sum(x*x)-n*mx*my
    
    b1 = cd/d
    b0 = my-b1*mx
    
    return (b0,b1)


def plot_reg_line(x,y,b) :
    plt.scatter(x,y,color="m",marker="o",s=30)
    y_pred = b[0] + b[1]*x
    plt.xlabel('x')
    plt.ylabel('y')
    plt.plot(x,y_pred,color='g')
    plt.show()
    
    
    
x = np.array([0,1,2,3])
y = np.array([1,3,2,5])

b = calc_coeff(x,y)
print('coeffiecients: ',b)
plot_reg_line(x,y,b)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

def cal_coeff(x,y) :
    n = np.size(x)
    mx = np.mean(x)
    my = np.mean(y)
    
    sxy = np.sum(y*x)-n*mx*my
    sxx = np.sum(x*x)-n*mx*my
    
    b1 = sxy/sxx
    b0 = my - b1*mx
    
    return (b0,b1)


def plot_line(x,y,b):
    plt.scatter(x,y,color='m',marker='o',s=30)
    y_pred = b[0] +b[1]*x
    plt.plot(x,y_pred)
    plt.xlabel('x')
    plt.ylabel('y')
    plt.show()
    
    

df = pd.read_csv('data.csv')    
x = df['X']
y = df['Y']

a = cal_coeff(x,y)

plot_line(x,y,a)
```
